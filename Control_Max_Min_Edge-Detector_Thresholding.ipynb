### This is a Desktop Application that apply "Canny Edge Detector" on an image that you will choose from your PC and gives you a control on max. and min. thresholding vlaue of Canny Edge Detector

#### For Tkinter Widgets  check this [LINK](https://www.dummies.com/programming/python/using-tkinter-widgets-in-python/)

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tkinter import *
from tkinter import filedialog
from PIL import ImageTk , Image
```

```python
# Build the Previous Tkinter App with fixed image size for all images and adjust Canny parametrs with scale widget and submit button

def controlmaxminedge():
    global panelA, panelB                   # global reference to the image panels
    path = filedialog.askopenfilename()     # open a file chooser dialog and allow the user to select an input image

    image = cv2.imread(path)               
    global gray
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    edged = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)         # Yeah pillow needs it in RGB
    
    # Resizing the original image and edged image.
    image = cv2.resize(image, (300, 200))
    edged = cv2.resize(edged, (300, 200))
   

    image = Image.fromarray(image)       # convert the images to PIL format...
    edged = Image.fromarray(edged)
    
    image = ImageTk.PhotoImage(image)    # ...and then to tkinter format
    edged = ImageTk.PhotoImage(edged)     

        
    # if the panels are None, initialize them
    if panelA is None or panelB is None:
        panelA = Label(image=image)         # the first panel will store our original image
        panelA.image = image
        panelA.pack(side="left", padx=10, pady=10)   # We need pack for localization
        
        panelB = Label(image=edged)          # while the second panel will store the edge map
        panelB.image = edged
        panelB.pack(side="right", padx=10, pady=10)
 
    else:                                # otherwise, update the image panels
        panelA.configure(image=image)
        panelB.configure(image=edged)
        panelA.image = image
        panelB.image = edged
            
            
# initialize the window toolkit along with the two image panels
root = Tk()
panelA = None
panelB = None


# create a button, then when pressed, will trigger a file chooser
# dialog and allow the user to select an input image; then add the
# button the GUI
btn = Button(root, text="Select an image", command=controlmaxminedge) # yourapp , text ,  binded function
btn.pack(side="bottom", fill="both", expand="yes", padx="10", pady="10") #  Localization

# Create a two sliders
# First one for min value of canny thresholding filter
slider1 = Scale(root, from_=0, to=255, orient= HORIZONTAL)
slider1.set(25)
slider1.pack(side="bottom", fill="both", expand="yes", padx="7", pady="7")
label1 = Label(root, text="Minimum Value")
label1.pack(side="bottom", fill="both", expand="yes", padx="7", pady="7")

# Second one for max value of canny thresholding filter
slider2 = Scale(root, from_=0, to=255, orient= HORIZONTAL)
slider2.set(25)
slider2.pack(side="bottom", fill="both", expand="yes", padx="5", pady="5")
label2 = Label(root, text="Maximumn Value")
label2.pack(side="bottom", fill="both", expand="yes", padx="5", pady="5")

def apply():
    
    minValue = slider1.get()
    maxValue = slider2.get()
    
    edged = cv2.Canny(gray, minValue, maxValue)
    edged = cv2.resize(edged, (300, 200))
    edged = Image.fromarray(edged)
    edged = ImageTk.PhotoImage(edged)
    
    panelB.configure(image=edged)
    panelB.image = edged
    
applybtn = Button(root, text="Apply", command=apply) # yourapp , text ,  binded function
applybtn.pack(side="bottom", fill="both", expand="yes", padx="3", pady="3") #  Localization
    
    
# kick off the GUI
root.mainloop()
```
